# Principal Component Analysis (PCA)

This notebook applies **Principal Component Analysis (PCA)** to the processed TCGA PANCAN gene expression dataset. The objective is to evaluate PCA as a linear dimensionality reduction technique for multiclass cancer classification.

Model performance is assessed using **Repeated Stratified 5-Fold Cross-Validation**  with **10 repetitions**. To prevent data leakage, feature standardization and PCA are fitted exclusively on the training data within each cross-validation fold.

For each training/test split, the training data are standardized, PCA is fitted on the standardized training data, and the resulting principal components are used to train a Logistic Regression classifier. The fitted scaler and PCA model are then applied to the corresponding test data before evaluation. The procedure is repeated for **20, 50, 100, and 200** principal components.


## Workflow

1. Load the processed dataset.
2. Import the required libraries.
3. Define the cross-validation strategy:
   - Repeated Stratified 5-Fold Cross-Validation (5 folds × 10 repetitions).
4. Define the numbers of principal components:
   - *n_components = [50, 100, 200]*
5. For each value of *n_components*:
   - For each cross-validation split:
     - Split the data into training and test sets.
     - Fit StandardScaler on the training data.
     - Transform the training and test data.
     - Fit PCA on the standardized training data.
     - Transform the training and test data.
     - Train the Logistic Regression classifier.
     - Generate predictions.
     - Store the evaluation metrics.
6. Aggregate the results across all cross-validation splits for each value of *n_components* by Computing their mean and standard deviations.

## 1. Data and Imports

```python
# Import required libraries
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.decomposition import PCA
import time

random_state = 42
```

```python
# Set project root
import sys
from pathlib import Path

project_root = Path("..").resolve()
sys.path.insert(0, str(project_root))
```

```python
# Import custom evaluation functions
from src.evaluation import evaluate_classification, summarize_results
```

```python
# Load the processed dataset
X = pd.read_csv(project_root / "data" / "processed" / "X_processed.csv")
X = X.iloc[:, 1:]
y = pd.read_csv(project_root / "data" / "processed" / "y_processed.csv",index_col=0).iloc[:, 0]
```

## 2. Define Experimental Settings

```python
# Cross-validation strategy
cv = RepeatedStratifiedKFold(n_splits=5, n_repeats=10, random_state=random_state)
```

```python
# Numbers of principal components to evaluate
n_components_list = [20, 50, 100, 200]
```

```python
# Classifier
classifier = LogisticRegression(max_iter=1000, random_state=random_state)
```

```python
import time
from sklearn.decomposition import SparsePCA
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler

# get one representative fold, same way your CV loop does it
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=random_state)
train_idx, test_idx = next(skf.split(X, y))

X_train = X.iloc[train_idx]

# scale, same as inside your PCA loop
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

# time a single Sparse PCA fit at k=200
spca = SparsePCA(n_components=200, alpha=1.0, random_state=random_state, max_iter=1000)

start = time.perf_counter()
spca.fit(X_train_scaled)
elapsed = time.perf_counter() - start

print(f"Single Sparse PCA fit at k=200: {elapsed:.1f} seconds")
print(f"Estimated full CV cost (50 folds × 4 k-values, k=200 as upper bound): "
      f"{elapsed * 50 * 4 / 60:.1f} minutes (lower bound; smaller k may be faster)")
```

## 3. PCA Classification Pipeline

```python
# Start total timer
total_start = time.perf_counter()

# Initialize result storage

# CLassification Performance
classification_results = {}
classification_summary = {}

# Representation Quality
representation_results = {}
representation_summary = {}

# Reconstruction
reconstruction_results = {}
reconstruction_summary = {}

# Runtime
runtimes = {}

# Evaluate PCA using repeated stratified cross-validation
for n_components in n_components_list:

    # Start timer for this value of n_components
    component_start = time.perf_counter()

    # Store results for all CV splits
    classification_split_results = []
    representation_split_results = []
    reconstruction_split_results = []

    # Cross-validation loop
    for train_idx, test_idx in cv.split(X, y):

        # Split data
        X_train = X.iloc[train_idx]
        X_test = X.iloc[test_idx]

        y_train = y.iloc[train_idx]
        y_test = y.iloc[test_idx]

        # Standardize
        scaler = StandardScaler()

        X_train_scaled = scaler.fit_transform(X_train)
        X_test_scaled = scaler.transform(X_test)

        # PCA
        pca = PCA(n_components=n_components, random_state=random_state)

        X_train_pca = pca.fit_transform(X_train_scaled)
        X_test_pca = pca.transform(X_test_scaled)

        # Classification

        # Classifier
        classifier.fit(X_train_pca, y_train)

        # Predict
        y_pred = classifier.predict(X_test_pca)

        # Evaluate
        classification_split_results.append(evaluate_classification(y_test, y_pred))

        # Representation
        
        representation_split_results.append({
            "Explained Variance":
                pca.explained_variance_ratio_.sum()
        })

        # Reconstruction

        X_train_reconstructed = pca.inverse_transform(X_train_pca)
        X_test_reconstructed = pca.inverse_transform(X_test_pca)

        X_train_reconstructed = scaler.inverse_transform(X_train_reconstructed)
        X_test_reconstructed = scaler.inverse_transform(X_test_reconstructed)

# Compute reconstruction metrics
        
    # Store classification results
    classification_results[n_components] = classification_split_results

    classification_summary[n_components] = summarize_results(
        classification_split_results
    )

    # Store representation results
    representation_results[n_components] = representation_split_results

    representation_summary[n_components] = summarize_results(
        representation_split_results
    )

    # Store reconstruction results
    reconstruction_results[n_components] = reconstruction_split_results
    
    # End timer for this value
    component_end = time.perf_counter()

    # Store runtime
    runtimes[n_components] = component_end - component_start
    
    print(
    f"PCA ({n_components:>3} components): "
    f"{runtimes[n_components]:.2f} seconds"
)

# End total timer
total_end = time.perf_counter()

print(f"\nTotal runtime: {(total_end - total_start)/60:.2f} minutes")
```

```python
# Summarize classification performance
for n_components, summary in summary_results.items():
    print(f"\nPCA ({n_components} components)")
    display(summary)
```
